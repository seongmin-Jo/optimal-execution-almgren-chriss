```python
# market impact ftn

def function_g(v, gamma, alpha):
    """
    Permanent Market Impact
    g(v) = gamma * v^alpha
    
    v = nk/tau : trading rate 
    gamma : Permanent Impact Constant
    """
    return gamma * (v ** alpha)

def function_h(v, tau, eta, epsilon, beta):
    """
    Temporary Impact
    h(v) = epsilon * singn(nk) + eta*(v)
    
    v = nk/tau : trading rate 
    eta : Price Impact for Each 1% of Daily Volume Traded
    epsilon : Fixed Cost of Selling
    """
    return epsilon * np.sign(v * tau) + eta * v
```
