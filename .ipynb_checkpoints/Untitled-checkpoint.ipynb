# Price Dynamics

We will assume that the stock price evolves according to a discrete arithmetic random walk:
$$
S_k=S_{k-1}+\sigma \tau^{1 / 2} \xi_k
$$
for $k=1, \ldots, N$ and where:
$$
S_k=\text { stock price at time } k
$$
